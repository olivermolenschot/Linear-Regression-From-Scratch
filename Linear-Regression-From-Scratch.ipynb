a) Give the critical values b and w for J (the critical values are the zeros of the first

```python
%%latex
In order to find the critical values of b and w, 
we must first find the partial derivatives of J with respect to the previously mentioned variables:
$$
\frac{\partial J}{\partial w} = 1\div2m*\sum_{i=1}^{m} [2*(wx_i+b - y_i)*x_i]
\\\frac{\partial J}{\partial w} = 1\div m*\sum_{i=1}^{m} (wx_i^2+bx_i-y_ix_i)$$
By separating the summation into three distinct summations, we get:
$$
\frac{\partial J}{\partial w} = w\div m*\sum_{i=1}^{m}(x_i^2) + b\div m*\sum_{i=1}^{m}(x_i) - 1\div m*\sum_{i=1}^{m} (y_ix_i)$$
Using the notation for ${\overline{x}}$ to represent the mean, we can simplify the equation:
$$
\frac{\partial J}{\partial w} = w\overline{x^2}+b\overline{x}-\overline{xy}$$
Setting the partial equal to 0, we arrive at the following:
$$
\frac{\partial J}{\partial w} = 0$$
\begin{equation}
\label{eu_eqn}
w\overline{x^2}+b\overline{x}-\overline{xy} = 0
\end{equation}
For now, we will leave this equation as is and find the critical value of b
$$\frac{\partial J}{\partial b} = 1\div2m*\sum_{i=1}^{m} [2*(wx_i+b - y_i)*1]\\
\frac{\partial J}{\partial b} = w\div m*\sum_{i=1}^{m}(x_i) + b*m\div m-1\div m*\sum_{i=1}^{m} (y_i)
\\\frac{\partial J}{\partial b} = w\overline{x}+b-\overline{y}$$
By setting the partial derivative equal to 0, we can solve for b:
$$
w\overline{x}+b-\overline{y}=0 \\
b = \overline{y}-w\overline{x} 
$$

Substituting the value of b within $\frac{\partial J}{\partial w}$, we derive an equation where w is solely dependent on values related to x and y: 
$$
w\overline{x^2}+(\overline{y}-w\overline{x})\overline{x}-\overline{xy} = 0
\\ w(\overline{x^2}-(\overline{x})^2)= \overline{xy}-\overline{x}*\overline{y}
\\ w=\frac{\overline{xy}-\overline{x}*\overline{y}}{\overline{x^2}-(\overline{x})^2}
$$
Thus, equation 2 and 3 represent the critical values of b and w for J, respectively.
```

```python
## The next functions presented are most of the functions that I will use to calculate the hypothesis function, loss function 
## and gradient descents. 
```

```python
def y_prediction(x,w,b):
    return w*x+b
```

```python
def loss_function(dataset,w,b):
    summation = 0
    for i in range(1,dataset.shape[0]+1):
        summation += (dataset['income'][i]*w+b-dataset['happiness'][i])**2
    return summation/(2*dataset.shape[0])
```

```python
def partial_deriv_J_w(dataset,w,b):
    return w*calculate_x_bar(dataset,2)+b*calculate_x_bar(dataset,1)-calculate_xy_bar(dataset)
    
```

```python
%%latex
$\frac{\partial J}{\partial w} = w\overline{x^2}+b\overline{x}-\overline{xy}$
```

```python
def partial_deriv_J_b(dataset,w,b):
    return w*calculate_x_bar(dataset,1)+b-calculate_y_bar(dataset)
```

```python
%%latex
$\frac{\partial J}{\partial b} = w\overline{x}+b-\overline{y}$
```

```python
def calculate_x_bar(dataset,power):
    output =0
    for i in range(1,dataset.shape[0]+1):
        output += dataset['income'][i]**power
    output = output/dataset.shape[0]
    return output
```

```python
def calculate_y_bar(dataset):
    output =0
    for i in range(1,dataset.shape[0]+1):
        output += dataset['happiness'][i]
    output = output/dataset.shape[0]
    return output
```

```python
def calculate_xy_bar(dataset):
    output =0
    for i in range(1,dataset.shape[0]+1):
        output += dataset['income'][i]*dataset['happiness'][i]
    output = output/dataset.shape[0]    
    return output
```

```python
def prediction_array(dataset,x,w,b):
    prediction_list = []
    for i in range(1,dataset.shape[0]+1):
        prediction_list.append(y_prediction(dataset['income'][i],w,b))
    return prediction_list
```

```python
def special_scatterplot(dataset,x,w,b):
    prediction_list = prediction_array(dataset,x,w,b)
    plt.scatter(income,happiness)
    plt.xlabel('income')
    plt.ylabel('happiness')
    plt.plot(dataset['income'],prediction_list,color='yellow',linewidth=3)
    plt.show()
```

```python
def batch_gradient(dataset,alpha,convergence_criteria,iterations):
    w = -100
    b = 1
    cost_batch = []
    cost_batch.append(loss_function(dataset,w,b))
    index = 0
    for i in range(0,iterations):
        w = w - alpha*partial_deriv_J_w(dataset,w,b)
        b = b - alpha*partial_deriv_J_b(dataset,w,b)
        cost_batch.append(loss_function(dataset,w,b))
        if (abs(cost_batch[index+1]-cost_batch[index])<convergence_criteria):
            return loss_function(dataset,w,b)
        else:
            index += 1
    return loss_function(dataset,w,b)
```

```python
def stochastic_gradient(dataset,alpha,convergence_criteria,iterations):
    w = -100
    b = 1
    cost_stochastic = []
    cost_stochastic.append(loss_function(dataset,w,b))
    index = 0
    counter_for_data = 1 ## this helps iterate over the elements in the dataset
    for i in range(0,iterations):
        ## this makes sure to stay within bounds of the dataset
        if (counter_for_data==dataset.shape[0]+1):
            counter_for_data = 1
        x = dataset['income'][counter_for_data]
        y = dataset['happiness'][counter_for_data]
        w = w - alpha*stochastic_partial_deriv_J_w(w,b,x,y)
        b = b - alpha*stochastic_partial_deriv_J_b(w,b,x,y)
        cost_stochastic.append(stochastic_loss_function(w,b,x,y))
        if (abs(cost_stochastic[index+1]-cost_stochastic[index])<convergence_criteria):
            return loss_function(dataset,w,b)
        else:
            index += 1
        counter_for_data +=1


    return loss_function(dataset,w,b)
```

```python
def mini_batch_gradient(dataset,alpha,convergence_criteria,iterations,batch_size):
    list_batch_size = [0]
    batches = int(dataset.shape[0]/batch_size)
    ## this checks for the case where the batch size is not a divisor of the number of inputs within the data
    # In greater detail, it creates indeces that will later be used to split the dataset into batches
    for i in range(0,batches):
        list_batch_size.append(list_batch_size[i]+batch_size)
    
    if (dataset.shape[0]-(batches*batch_size)>0):
        list_batch_size.append(list_batch_size[i+1]+(dataset.shape[0])-(batches*batch_size))
    w = -100
    b = 1
    cost_mini_batch = []
    cost_mini_batch.append(loss_function(dataset,w,b))
    index = 0
    iteration_batch = 0
    
    for i in range(0,iterations):
        ## the two lines below allow me to iterate over the dataset, 50 inputs at a time, by creating a deep copy. Modulo 11 ensures
        ## that the code stays within bounds.
        iteration_batch = iteration_batch%10
        batch_copy = dataset[list_batch_size[iteration_batch]:list_batch_size[iteration_batch+1]].copy()
        batch_copy = batch_copy.reset_index(drop=True)
        batch_copy.index = range(1, len(batch_copy) + 1)
        w = w - alpha*partial_deriv_J_w(batch_copy,w,b)
        b = b - alpha*partial_deriv_J_b(batch_copy,w,b)
        cost_mini_batch.append(loss_function(dataset,w,b))
        if ((abs(cost_mini_batch[index+1]-cost_mini_batch[index])<convergence_criteria)):
            break
        else:
            index += 1
        iteration_batch +=1
    return loss_function(dataset,w,b)
```
